```python
#1. problem statement is Predict the Chronic Kidney disease based on several parameters
#2. Dataset has 25 columns and 399 rows
#3. For standardize the dataset preprocessing is require - get_dummies, standardscaler
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv('CKD.csv')
dataset # column C, F, G, H, I, S, T, U, V, W, X, Y has nominal data
#column Y has output rest of them all are input
```

```python
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
dataset #convert the nominal data into ordinal data
```

```python
#seperate input and output
dataset.columns
```

```python
input = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
output = dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (input, output, test_size = 1/3, random_state = 0)
```

```python
dataset [['classification_yes']].value_counts()#dataset is highly imbalanced
#dataset has 2 class class 1 = yes, class 0 = no
```

```python
from sklearn.preprocessing import StandardScaler
sc =  StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
parameter = {  'loss': [ 'log_loss', 'modified_huber','squared_error', 
                        'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
    'penalty': ['l2'],
    'alpha': [0.0001],
    'l1_ratio': [0.15],
    'fit_intercept': [True],
    'max_iter': [1000],
    'tol': [0.001],
    'shuffle': [True],
    'verbose': [0],
    'epsilon': [0.1],
    'n_jobs': [None],
    'random_state': [None],
    'learning_rate': ['optimal'],
    'eta0': [0.01],
    'power_t': [0.5],
    'early_stopping': [False],
    'validation_fraction': [0.1],
    'n_iter_no_change': [5],
    'class_weight': [None],
    'warm_start': [False],
    'average': [False]
    }
grid = GridSearchCV(SGDClassifier(), parameter, refit = True, verbose = 3, n_jobs = -1, scoring = 'f1_weighted')
grid.fit(x_train, y_train)
```

```python
result = grid.cv_results_
grid_pred = grid.predict(x_test)

#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_pred)
print (cm)

from sklearn.metrics import classification_report
clf = classification_report(y_test, grid_pred) 
print(clf)
```

```python
#generate f1 score value
from sklearn.metrics import f1_score # to get the f1 score value
f1_macro = f1_score(y_test, grid_pred, average = 'weighted')
print("The f1_macro value for best parameter {}.". format(grid.best_params_), f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc = roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
roc
```
